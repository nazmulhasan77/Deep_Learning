```python
import os
import cv2
import numpy as np
from PIL import Image
import pillow_heif

# Enable HEIC support
pillow_heif.register_heif_opener()

input_folder = "flowers1"
output_folder = "flowers"

os.makedirs(output_folder, exist_ok=True)

valid_exts = ('.jpg', '.jpeg', '.png', '.bmp', '.heic')

count = 0

for filename in os.listdir(input_folder):
    if filename.lower().endswith(valid_exts):
        img_path = os.path.join(input_folder, filename)

        # 🔥 Handle HEIC separately
        if filename.lower().endswith('.heic'):
            try:
                pil_img = Image.open(img_path).convert("RGB")
                img = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
            except Exception as e:
                print(f"Skipping HEIC: {filename} | Error: {e}")
                continue
        else:
            img = cv2.imread(img_path)

        if img is None:
            print(f"Skipping: {filename}")
            continue

        # Resize
        resized = cv2.resize(img, (224, 224))

        # Save as JPG
        new_name = os.path.splitext(filename)[0] + ".jpg"
        save_path = os.path.join(output_folder, new_name)

        cv2.imwrite(save_path, resized, [int(cv2.IMWRITE_JPEG_QUALITY), 95])

        count += 1

print(f"Done! {count} images processed.")
```
